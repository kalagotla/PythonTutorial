# Problem 1: Plot Bessel function

```python
# modules import
import numpy as np
from scipy.special import jn
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4*np.pi, 100) # x-variable
```

```python
for i in range(7):
    plt.plot(x, jn(i, x)) # Plot bessel function in multiple orders
```
