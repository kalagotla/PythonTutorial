```python
# modules import
import numpy as np
from scipy.special import jn
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4*np.pi, 100) # x-variable
```

```python
for i in range(7):
    plt.plot(x, jn(i, x)) # Plot bessel function in multiple orders
```

# Is there an easy way to catch an error without typing try and except all the time?

```python
def error_catch(func, input_string, error_statement):
    # A function to catch user input errors
    
    while True:
        try:
            val = func(input(input_string))
            break
        except ValueError:
            print(error_statement)
    
    return val

# Instead of entering input command, a = input('integer value: '); Use
a = error_catch(float, 'float value: ',
               'Please enter a float')

print(a+5)
```
